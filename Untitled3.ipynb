```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
```

```python
invalid = ['a/','b/','c/']
csv_file = pd.read_csv('value_agri.csv',na_values=invalid)
```

```python
df = pd.DataFrame(csv_file)
```

```python
df.rename(columns={ 'Geolocation': 'geolocation', 'Subsector': 'subsector', 'Year': 'year', 'Value of Production in Agriculture': 'value' }, inplace=True)
```

```python
len(df)
df.head()
```

```python
df['ln_year'] = pd.Series(np.log(df['year']), df.index)
df['ln_value'] = pd.Series(np.log(df['value']), df.index)
```

```python
subset = df[(df['geolocation']=='PHILIPPINES') & (df['subsector']=='TOTAL')]
```

```python
subset.describe()
```

```python
subset.sample(22)
```

```python
subset.corr()
```

```python
result = sm.ols(formula='ln_value ~ ln_year', data=subset).fit()
```

```python
print(result.summary())
```

```python
ypred = result.predict(subset['ln_year'])
```

```python
print(ypred)
```

```python
subset.plot(kind='line', x='year', y='ln_value', title='Value per Year')
```
